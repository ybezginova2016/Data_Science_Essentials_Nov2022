# Machine Learning Algorithms

***Machine Learning*** is the process of deploying compute power to identify trends and patterns in data in order to make predictions about how that data will behave in the future.

***Machine Learning (ML)*** is a field of computer science where algorithms learn from data. ML is a combination of ***Computational Statistics, Pattern Regression, Mathematical Optimization, Computational learning theory in artificial intelligence.***

### Machine Learning Use Cases

### Machine Learning models can be used to:
1) identify relationships between variables (***regression***)

2) assign new data to the correct group (***classification***)

3) find the distribution of inputs in a certain space (***density estmation***)

4) simplify inputs by mapping them into a lower dimensional space (***dimensionality reduction***)

5) group similar data points together, without knowing the groups beforehand (***clustering***)

### Machine Learning Algorithms by Function

#### SUPERVISED LEARNING

- **Regression (labeled)**
    - Linear Regression
    - Logistic Regression
    
- **Instance-based classification (labeled)** learns from labeled data to predict categories for new data points; the class labels are discrete, unordered values:
    - K-nearest neighbor algorithm (KNN)
    
#### UNSUPERVISED LEARNING
    
- **Dimension reduction (unlabeled, unsupervised learning)** takes unlabeled, continuous, high-dimensional data and compresses it into a smaller synthetic representation:
    - Explanatory factor analysis
    - Principal component analysis (PCA)
    
    
- **Clustering (unlabeled)** takes unlabeled, categorical data and organizes it into clusters based on similarity in attribute values of the data points:
    - **K-means clustering**
        - Location-based crime prediction and prevention
        - Customer segmentation for marketing
        - Document classification
        - Insurance fraud prediction and prevention
        - IT operational efficiences
        - Oil and gas site evaluations
        - Video analytics
        - Patients segmentation for preventative medicine
    - Hierarchical methods
    - DBSCAN

    
- **Association rules**
    - Association rules with Apriori
    
- **Deep learning**
    - Perceptron
    
- **Decision trees**
    - CART
    
- **Bayesian statistics**
    - Naive Bayes
    
- **Ensemble**
    - Random Forest
    
- **Regularization** (_used in neural networks_)
    - https://machinelearningmastery.ru/concept-of-regularization-28f593cf9f8c/

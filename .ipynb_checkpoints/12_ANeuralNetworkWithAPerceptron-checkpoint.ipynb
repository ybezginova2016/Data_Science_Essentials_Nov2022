## Fundamental Concepts

### Perceptron
***A perceptron*** is a neural network with just one layer. It is a linear classifier that outputs a binary responsive variable. Consequently, the algorithm is called a ***'linear binary classifier'***.

#### Linear separability
- Data is said to have 'linear separability' if it can be cleanly classified into one of two classes.
- Your data must be linearly separable in order for a perceptron to operate properly.

### Four Essential Elements of a Perceptron

#### 1) An input layer
#### 2) Weights and bias
#### 3) A weighted sum

This is a single-layer neural network. It generates a simple linear model of a phenomenon. Basically, a perceptron is just a single-layer neural network. The neural network shown here deploys a simple linear activation function on a network that has only one layer. Consequently, the network is acting as a linear model. 



#### 4) An activation function
- A mathematical function that is deployed on each unit in a neural network.

- All units in a shared layer deploy the same activation function. 

- The purpose of activation functions is to enable neural networks to model complex, nonlinear phenomenon. 

There are many types of activation functions that you can use, including:

- ***linear activation function***: tf.matmul() - single layer perception
- ***logistic sigmoid function***: use these often in the final output layer, useful with binary input features
- ***threshold function***: useful with binary features
- ***rectified linear unit (ReLU)***
- ***SoftMax***

```python
import numpy as np
import pandas as pd
import sklearn

from pandas import Series, DataFrame
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.linear_model import Perceptron
```

```python
iris = datasets.load_iris()

X = iris.data
y = iris.target

X[0:10,]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
standardize = StandardScaler()

standardized_X_test = standardize.fit_transform(X_test)

standardized_X_train = standardize.fit_transform(X_train)
```

```python
standardized_X_test[0:10,]
```

```python
perceptron = Perceptron(max_iter=50, eta0=0.15, tol=1e-3, random_state=15)

perceptron.fit(standardized_X_train, y_train.ravel())
```

```python
y_pred = perceptron.predict(standardized_X_test)
```

```python
print(y_test)
```

```python
print(y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

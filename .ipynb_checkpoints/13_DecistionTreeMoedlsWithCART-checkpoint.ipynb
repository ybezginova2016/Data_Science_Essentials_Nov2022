# Decision Trees With CART (Classification and Regression)

## Decision Tree
**A decision-support tool that models decisions in order to predict probable outcomes of those decisions.**

### Decision Trees Have Three Types of Nodes
- **Root node** is node that represents the entire population or sample, and this (mumbles) gets divided into two or more homogenous sets. 

- **Decision node** is where a subnode splits into further subnodes.

- **Leaf node** is just the end node or the furthest subnode. It doesn't split.



## Decision Tree Algorithms
**A class of supervised machine learning methods that are useful for making predictions from non-linear data.**

Decision tree algorithms are an appropriate fit for:
- ***Continuous*** input and / or output variables.
- ***Categorical*** input and / or output variables.

### Two Types of Decision Trees

#### 1 - Categorical variable decision tree (or ***classiication trees***):
when you use a decision tree to predict for a categorical target variable.

#### 2 - Continuous variable decision tree (or ***regression trees***):
when you use a decision tree to predict for a continuous target variable.


### Main Benefits of Decision Trees Algorithms
- Decision trees are nonlinear models, hence they can fit the data even if it's nonlinear.

- Decision trees are very easy to implement. It's often said that decision trees mirror human thinking and hence, they have better interpretability than other regression and classification methods. 

- Decision trees can be graphically represented in a way that's easy to read and easy to interpret. 

- Decision trees require less data preparation. It's also nice that they can take care of categorical variables without having to implement one-hot encoding.
